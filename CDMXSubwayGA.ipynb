```python

def generate_path(G, start_node, end_node):
    # Check if the start and end nodes are in the graph
    if start_node not in G or end_node not in G:
        raise ValueError("Start or end node not found in the graph.")
    # Initialize the path and cost
    current_node = start_node
    path = [current_node]
    cost = 0

     # Traverse the graph until the end node is reached
    while current_node != end_node:
        #Get the neighbors of the current node
        neighbors = list(G[current_node])

        # Remove any neighbors that have already been visited
        unvisited_neighbors = [n for n in neighbors if n not in path]

        if not unvisited_neighbors:
            return None  # Return None to indicate path generation failed
        # Choose a random unvisited neighbor
        next_node = random.choice(unvisited_neighbors)

        # Update the path and cost
        path.append(next_node)
        edge = G[current_node][next_node]
        cost += edge.get('hops', 1)  # Assuming default hop cost of 1 if not specified

        # Move to the next node
        current_node = next_node

    return path, cost
```

```python
#Create an individual(A path with its respective cost)
def create_individual(G, start_node, end_node):
    try:
        result = generate_path(G, start_node, end_node)
        if result is None:
            return None  # Skip adding this individual if no valid path was found
        path, cost = result
        return {'path': path, 'cost': cost}
    except ValueError as e:
        print(e)
        return None
```

```python
#Create a population
def initialize_population(G, start_node, end_node, population_size):
    population = []
    while len(population) < population_size:

        individual = create_individual(G, start_node, end_node)
        if individual:
            population.append(individual)
    return population
```

```python
#Determine fitness for an individual
def fitness(individual):
    if individual['cost'] == 0:
        return float('inf')  # Maximally good fitness for a direct connection.
    return 1 / individual['cost']
```

```python
#Calcualte fitness for a population
def evaluate_population(population):
    fitness_scores = [(individual, fitness(individual)) for individual in population]
    fitness_scores.sort(key=lambda x: x[1], reverse=True)  # Sort by fitness score, descending
    return fitness_scores

```

```python
# Use fitness proportionate selection to select the parents of the next generation with elitism to keep the best results
def selection(popRanked, eliteSize):
    selectionResults = []
    # Calculate the total fitness for normalization
    total_fitness = sum([fitness for _, fitness in popRanked])

    # Create a cumulative probability list
    cumulative_sum = 0
    cumulative_probs = []
    for index, (_, fitness_score) in enumerate(popRanked):
        cumulative_sum += fitness_score
        cumulative_probs.append((index, 100 * cumulative_sum / total_fitness))

    # Elite selection: append indices of the elites
    for i in range(eliteSize):
        selectionResults.append(i)  # Append index of elite individuals

    # Roulette wheel selection: append indices based on fitness proportion
    for _ in range(len(popRanked) - eliteSize):
        pick = 100 * random.random()
        for index, (cum_index, cum_prob) in enumerate(cumulative_probs):
            if pick <= cum_prob:
                selectionResults.append(cum_index)
                break

    return selectionResults
```

```python
#Create the mating pool for the next generation
def matingPool(population, selectionResults):
    matingpool = []
    for i in range(0, len(selectionResults)):
        index = selectionResults[i]
        matingpool.append(population[index])
    return matingpool
```

```python
def calculate_path_cost(G, path):
    total_cost = 0
    # Iterate through pairs of stations in the path
    for i in range(len(path) - 1):
        start_station = path[i]
        next_station = path[i + 1]
        
        # Check if the edge between the current station and the next station exists
        if G.has_edge(start_station, next_station):
            # Add the cost of the edge to the total cost
            total_cost += G[start_station][next_station].get('hops', 0)  # Default to 0 if no hops key found
        else:
            # If there's no direct link (which should be an error in path planning), you can handle it
            # For now, let's raise an exception
            raise ValueError(f"No direct connection between {start_station} and {next_station} found.")
    
    return total_cost
```

```python

def find_path(G, start, end):
    try:
        # Retrieves the shortest path between start and end nodes
        path = nx.shortest_path(G, start, end)
        return path if path[0] != start else path[1:]  # Avoid repeating the start if it's already added
    except nx.NetworkXNoPath:
        return []

def breed(parent1, parent2, G, start_node, end_node):
    # Extract paths from the dictionaries assuming the paths are stored under the key 'path'
    path1 = parent1['path']
    path2 = parent2['path']
     # Choose crossover points, ensuring they are within the bounds of the shortest path
    geneA = int(random.random() * (len(path1) - 1)) + 1
    geneB = int(random.random() * (len(path2) - 1)) + 1

    startGene = min(geneA, geneB)
    endGene = max(geneA, geneB)
     # Start with the fixed start node, and add a segment from the first parent's path
    child_path = [start_node]
    if G.has_edge(start_node, path1[startGene]):
        child_path += path1[startGene:endGene]
    else:
        path_segment = find_path(G, start_node, path1[startGene])
        if path_segment and path_segment[0] == child_path[-1]:  # Prevent duplication
            path_segment = path_segment[1:]
        child_path += path_segment
    # Fill in the remainder of the route from the second parent's path, ensuring each station connects properly   
    last_station = child_path[-1]
    added_end_node = False
    for station in path2[geneB:]:
        if station not in child_path and station != last_station and G.has_edge(last_station, station):
            child_path.append(station)
            last_station = station
            if last_station == end_node:
                added_end_node = True
                break
     # Ensure the end node is added if not already present            
    if not added_end_node:
        if G.has_edge(last_station, end_node):
            child_path.append(end_node)
        else:
            path_to_end = find_path(G, last_station, end_node)
            if path_to_end and path_to_end[0] == last_station:  # Again, prevent duplication
                path_to_end = path_to_end[1:]
            child_path += path_to_end
    # Create a dictionary with the new child path and calculate its cost
    child_individual = {
        'path': child_path,
        'cost': calculate_path_cost(G, child_path)
    }
    return child_individual
```

```python
#Breeding function for an entire population ensurring to keep the best individuals

def breedPopulation(matingpool, eliteSize, G, start_node, end_node):
    children = []
    length = len(matingpool) - eliteSize
    pool = random.sample(matingpool, len(matingpool))

    # Carry over the elites directly to the next generation
    for i in range(eliteSize):
        children.append(matingpool[i])

    # Generate new children by breeding individuals from the pool
    for i in range(length):
        # Select parents for breeding
        parent1 = pool[i]
        parent2 = pool[len(matingpool) - i - 1]
        # Breed and add the child to the new generation
        child = breed(parent1, parent2, G, start_node, end_node)
        children.append(child)

    return children
```

```python
def mutate(individual, G, mutation_rate, start_node, end_node):
    if random.random() < mutation_rate:
        path = individual['path']  # Work with the path list inside the dictionary
        mutation_type = random.choice(['swap', 'insert', 'remove'])
        if mutation_type == 'swap' and len(path) > 3:
            # Attempt to swap two stations in the path, excluding start and end nodes
            for _ in range(10):# Try multiple times to find a valid swap
                idx1, idx2 = random.sample(range(1, len(path) - 1), 2)
                # Check if swap is valid
                if G.has_edge(path[idx1 - 1], path[idx2]) and \
                   G.has_edge(path[idx2], path[idx1 + 1]):
                    path[idx1], path[idx2] = path[idx2], path[idx1]
                    break
        # Attempt to insert a random station in the path, excluding start and end positions
        elif mutation_type == 'insert' and len(G.nodes()) > 2:
            for _ in range(10):# Try multiple times to find a valid insertion
                new_station = random.choice(list(G.nodes()))
                if new_station != start_node and new_station != end_node:
                    insert_pos = random.randint(1, len(path) - 1)
                    # Check if insertion is valid
                    if G.has_edge(path[insert_pos - 1], new_station) and \
                       G.has_edge(new_station, path[insert_pos]):
                        path.insert(insert_pos, new_station)
                        break
        elif mutation_type == 'remove' and len(path) > 3:
            for _ in range(10):
                # Attempt to remove a random station from the path, but not start or end
                remove_pos = random.randint(1, len(path) - 2)
                 # Check if removal is valid
                if G.has_edge(path[remove_pos - 1], path[remove_pos + 1]):
                    path.pop(remove_pos)
                    break
    return individual  # Return the individual with possibly modified path
```

```python
#Apply mutation to a population
def mutate_population(population, G, mutation_rate, start_node, end_node):
    mutated_population = []
    for individual in population:
        mutated_individual = mutate(individual, G, mutation_rate, start_node, end_node)
        mutated_population.append(mutated_individual)
    return mutated_population
```

```python
#Repeat the process to generate new generations

def nextGeneration(currentGen, eliteSize, mutationRate, G, start_node, end_node):
    popRanked = evaluate_population(currentGen)
    selectionResults = selection(popRanked, eliteSize)
    matingpool = matingPool(currentGen, selectionResults)
    #print("Mating pool")
    #print(matingpool)
    children = breedPopulation(matingpool, eliteSize, G, start_node, end_node)
    #print("Children")
    #print(children)
    nextGeneration = mutate_population(children, G, mutationRate, start_node, end_node)
    #print("After mutation")
    #print(nextGeneration)
    return nextGeneration
```

```python
def geneticAlgorithm(graph, popSize, eliteSize, mutationRate, generations, start_node, end_node):
    print("Start")
    pop = initialize_population(graph, start_node, end_node, popSize)
    print(f"Generation: 0")
    print("Initial cost: " + str(min(individual['cost'] for individual in pop)))
    best_individual = min(pop, key=lambda x: x['cost'])
    print(f"Best path this generation: {best_individual['path']}")
    print(f"Best cost this generation: {best_individual['cost']}")

    for i in range(generations):
        print(f"Generation: {i + 1}")
        pop = nextGeneration(pop, eliteSize, mutationRate, graph, start_node, end_node)

        # Find the best individual in the current population
        best_individual = min(pop, key=lambda x: x['cost'])
        print(f"Best path this generation: {best_individual['path']}")
        print(f"Best cost this generation: {best_individual['cost']}")

    # Print the overall best at the end of all generations
    overall_best_individual = min(pop, key=lambda x: x['cost'])
    print("End")
    print("Best overall path: ", overall_best_individual['path'])
    print("Best overall cost: ", overall_best_individual['cost'])
    return overall_best_individual
```
